I will be writing a function to select highly correlated feature keeping the threshold at 0.7(70%). However if this threshold is changed immensely, we might see other highly correlated feature as output.

```python
# with the following function we can select highly correlated features

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

A small description to make readers understand some statistical concepts related to this case analysis. Our goal in this analysis is to find out a correlation between 'ctr' and 'product_tier'. What is the cause of this correlation? We already know if a 'product_tier' is 'Premium', it has a higher preference listing followed by 'Plus'. Now our goal is to find how the strength of association can be measured. Previously we observed this strength amongst continous variables in the dataset using a correlation matrix. There are different tests for different scenarios. Some of them will be defined in short below:

(a) Chi-squared: Used for comparison between variables that are only categorical.
(b) Correlation: Used for measuring relationship between variables that are only continous.
(c) T-test: Used for comparison between categorical and continous variables. 

It's easy to decide in our case we have to use T-test. But since we have more than one categorical variable (3 tiers) we will be using a special type of T-test called as Anova. This helps us to find if there is a significant difference between these groups and how significant is the difference. 

Some other more general concepts in short that readers might find helpful: 

Null Hypothesis: There is no difference between the two or more variants. They all perform equally.
Alternate Hypothesis: There is a difference between the two or more variants. 

When to accept or reject null hypothesis? This is done by using p-value denoted normally as Î± (alpha), also known as significance value. If p-value â‰¤ 0.05, we reject Null hypothesis. This will tell if there is a significant difference in the variants. Let's proceed with our Anova test. I will be using scipy library to calculate p-value.


```python
#anova test(f-test)

a_test = as24[['product_tier', 'ctr']]
```

```python
a_test.head(10)
```

```python
grps= pd.unique(a_test.product_tier.values)
```

```python
grps
```

```python
atest_data = {grp:a_test['ctr'][a_test.product_tier == grp] for grp in grps}
```

```python
atest_data
```

```python
import scipy.stats as stats
```

```python
F, p = stats.f_oneway(atest_data['Basic'], atest_data['Premium'], atest_data['Plus'])
```

You may question what is F and p here? p is p-value defined earlier above and F is the ratio of the variance of group means and means of within group differences. In our cases groups are the tiers.

```python
print(p)
```

```python
if p<0.05:
    print("reject null hypothesis") ##Rejecting null hypothesis means there is a statistical significant difference
else:
    print("accept null hypothesis") ##Accepting null hypothesis means there is a statistical significant difference  
```

```python
print(F)
```

So now that we know there is a statistical significant difference between the tiers. Our goal should be to measure how significant these differences are. This can be done by different methods, in our case we will be using Post-hoc test. A Post-hoc test is perfomed when we know that we are rejecting the null hypothesis (anova is significant) and there are three or more group variants. Post-hoc tests are of different types, e.g. Bonferroni method, Tukey HSD (Honestly significant difference), Games-Howell. 

When to use which type? 
Usually the most safest option considered is to use Bonferroni method. However no test is perfect, Bonferroni being a very conservative method often increases the Type-2 error ( We miss a significant that is truly there, i.e., less precision). 
For unequal sample sizes, better methods like Gabriel's (small sample) and Hochberg's GT2 (large sample) are used.
For unequal variances, we use Games-Howell method. There are many other methods that exist, I have just mentioned few of them as examples here.

Dont worry! we will be using more than one type just to compare. If our result is True with one method it should most likely be True with other method as well, However there might be a slight difference in significance values.

Let's first quickly check if our product tiers have unequal variances. It's not that hard to answer that they do have unequal variances, but lets perform a test to be sure. This can be done using Levene F-test. If p-value is less than 0.05 (significant), we can say the variances are not equal.

```python
##Posthoc test

##levene test to check if the product tiers have equal variances

myproduct = as24['product_tier']
myctr = as24['ctr']
```

```python
##Boolean creation
mytier1 = myproduct == 'Basic'
mytier2 = myproduct == 'Plus'
mytier3 = myproduct == 'Premium'
```

```python
##creating list using boolean

mytier1score = myctr[mytier1].dropna()
mytier2score = myctr[mytier2].dropna()
mytier3score = myctr[mytier3].dropna()
```

```python
from scipy.stats import levene 
```

```python
levene(mytier1score, mytier2score, mytier3score, center='mean')
```

Since the value of p is really less than 0.05 we can say the variances are not equal. When the variances are not equal we can use Games-Howell method as a type of Posthoc test. However Games-Howell is sometimes imprecise, pairwise comparison test. But we will be checking with all the other methods as well. Let's proceed.

```python
import pingouin as pg
from pingouin import pairwise_ttests
```

```python
##arguments in pairwise_ttests 
##dv = dependent variable in our case 'ctr'
##between = independent variable in our case 'product_tier'
##data = name of the dataframe
##padjust = the adjustment method which takes the following options : 'none', 'bonf', 
##'sidak', 'holm', 'fdr_bh' and 'fdr_by'. none means no correction, 'bonf' means bonferroni type of posthoc, 'fdr_bh' means Benjamini/Hochberg..etc
##But as we have unequal variances we will be using the Games-Howell. However we can try all of them.

##Games-Howell method uses a function 'pairwise_gameshowell' for padjust
```

Bonferroni is considered the safest option amongst all Post hoc tests.

```python
posthocs= pairwise_ttests(dv= 'ctr', between= 'product_tier', padjust = 'bonf', data= as24, correction = False)
posthocs
```

We can see p-corr as our p-value being the most significant between the variants: 'Basic' and 'Premium' followed by a very much smaller significant difference between the variants: 'Plus' and 'Premium' and even smaller significant difference between the variants: 'Basic' and 'Plus'. Note that the smaller the p-value, the bigger is the difference between the variants. The smallest p-value by a huge margin is noticed between the variants: 'Basic' and 'Premium', followed by 'Plus' and 'Premium' and least difference being between the variants: 'Basic' and 'Plus'. Let's plot this result and check if other methods give us the same result!
